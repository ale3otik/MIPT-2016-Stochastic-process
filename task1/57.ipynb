```python
# delete dummy nodes (with unknown gender and has no children(leaf in generation-tree))
def clear_data(processes) :
    processes_new = []
    for proc in processes :
        new_proc = BranchingProcess()
        new_proc.generations = []
        children = set()
        proc.generations.reverse()
        for generation in proc.generations:
            new_gen = []
            for person in generation :
                #delete no-genders leaves
                if person.gender == 'male' or person.gender == 'female':
                    children.add(person.name)
                    new_gen.append(person)   
                else :     
                    is_leaf = True
                    for ch in person.children:
                        if ch in children :
                            is_leaf = False
                            break
                    if not is_leaf :
                        print("Warning! empty-no-leaf")
                        new_gen.append(person)   
    
            if len(new_gen) > 0 :
                new_proc.generations.append(new_gen)

        if len(new_proc.generations) > 1 :
            new_proc.generations.reverse()
            processes_new.append(new_proc)
    return processes_new
```

```python
def pois_likelyhood_estimation(X) :
    if len(X) > 0:
        return np.mean(X)    
    else :
        return -10
```

В данных сложные фамилии могут быть представлены двумя способами, с пробелом или нижним подчеркиванием.

Поэтому нормализуем все имена в процессе обработки.

Также не будем рассматривать как бездетных мужчин, у которых не указана дата смерти, но которых уже "почти наверное" нет в живых ( > 96 лет), так как о них, вероятнее всего, просто утеряна информация. ((*) - критерий)

```python

def normalize_name(s) :
    return s.replace(' ' , '_')


#  add only non-leaf(not last generation in family) and older than 1950 birth year
#  delete if p has no death-year(but he must be already died)
def death_crit(person):
    return False #get_year(person.deathdate) <= -5000 and get_year(person.birthday) <= 1920
def is_bad_example(person) :
    return get_year(person.birthday) <= -5000 or get_year(person.birthday) >= 1950 
#         or death_crit(person)
            

X = []
# print(processes[1].generations[0][0])
t = 0
for proc in processes :
    offspring = dict()
    all_p = set()
    for generation in proc.generations:
        for person in generation :
            if person.gender == 'male':
                all_p.add(normalize_name(person.name))
    # calc number of sons for each father
    for generation in proc.generations:
        for person in generation :
            person.name = normalize_name(person.name)
            if person.gender == 'male' and len(person.parents) > 0 and len(person.parents[0]) > 0 :
                person.parents[0] = normalize_name(person.parents[0])
                if person.parents[0] not in all_p :
                    continue
                if person.parents[0] not in offspring :
                    offspring[person.parents[0]] = 0 
                offspring[person.parents[0]] += 1
    
    #assign by zero for no-sons fathers       
    i = 0
    for generation in proc.generations:
        i += 1
        for person in generation :
            person.name = normalize_name(person.name)
            if person.gender == 'male' and person.name not in offspring :
                if i < len(proc.generations) and not is_bad_example(person):
                    offspring[person.name] = 0
    vals = list(offspring.values())
    if len(vals) > 0 :
        X.append(vals)
    else:
        X.append([])
        
X_all = []
for x in X :
    X_all += x
```

 - для Пуассоновского распределения : $ \hat \lambda = \overline X$
 - для геометрического : $ \hat p = \frac{1}{\overline X + 1}$


```python
pois_lmb_all = pois_likelyhood_estimation(X_all)
pois_lmb = np.array([pois_likelyhood_estimation(x) for x in X])
p_geom = 1/(pois_lmb + 1)
p_geom_all = 1/(pois_lmb_all + 1)
print('lambda_all = ' + str(pois_lmb_all)) 
print('p_geom_all = ' + str(p_geom_all))
```

Критерий $\chi^2$ :

```python
def get_chisquare_check(prob, f_obs, mu) :
    n = np.sum(f_obs)
    fexp = [n*prob(i,mu) for i in range(len(f_obs))]
    val , p = sps.chisquare(f_obs, f_exp=fexp)
    return val,p
```

 - Допущение: Наш анализ не будет учитывать возможность добавления потомков к мужчинам, имеющимся в наших данных, но у которых не известны потомки. Так как таких не много, это не повлияет на результат.

 - Метод: разобьем всю интересующую нас временную ось на промежутки длины равной среднему расстоянию между поколениями, и для каждого человека найдем промежуток в котрый он попал. Разделим всех людей на "активных" и "не оактивных". Активные - те кто являются корнем своего процесса в данном поколении, значит нужно найти им родителя из предыдущего поколения. Пусть в поколении $i$ у нас $a_i$ активных людей, тогда добавим в поколение $i-1$ добавим $n_i = max\{n:\xi_1 + \dots + \xi_n \geq a_i \}$ активных людей, сгенерировав, соответственно, величины $\{\xi_i\}$.

 - Примечание: пусть количество людей в поколении $i$ - $p_i$ , а величина $c_l = avg\_life\_time/avg\_gen\_dist$ - среднее количество людей живущих во временном промежутке, относящимся к данному поколению. Тогда по оси OY будем откладывать величину $p_i * c_l$

```python
distr = [lambda lmb,sz : sps.poisson.rvs(lmb, size=sz),
         lambda p,sz : np.array(sps.geom.rvs(p,size=sz))-1 , 
         lambda p,sz: empirical_distr_all.rvs(size=sz)]
names = ['poiss', 'geom' , 'empirical']
params = [pois_lmb_all , p_geom_all , 0]
```

Найдем среднее время до появления следующего поколения

```python
gen_dists = []
for proc in processes:
    avg_byears = []
    for gen in proc.generations:
        bdays =[]
        for p in gen:
            if(p.gender != 'male'):
                continue
            bdy = get_year(p.birthday)
            if(bdy > -5000) :
                bdays.append(bdy)
        if(len(bdays) > 0):
            avg_byears.append(np.mean(bdays))
    for i in range(1,len(avg_byears)):
        if(avg_byears[i-1] - avg_byears[i] > 0) :
            gen_dists.append(avg_byears[i-1] - avg_byears[i])
# print(gen_dists)
avg_gen_distance = int(np.mean(gen_dists))
print('average year distance between generations = ' + str(avg_gen_distance))
```

Среднее число поколений живущих в данный отрезок времени:

```python
life_coef = avg_life_time / avg_gen_distance
print(life_coef)
```

```python
step = avg_gen_distance
def calc_backward(distrs,names,params,start,xlim,ylim):
    colors = ['r','g','b']
    start_time = start
    length = max_birthyear - start_time
    
    plt.figure(figsize=(12,6))
    plt.grid(True)
    plt.xlabel('years')
    plt.ylabel('population')
#     plt.xlim(start_time,start_time + length)
    plt.xlim(xlim[0],xlim[1])
    if ylim != [0,0]:
        plt.ylim(ylim[0],ylim[1])
    for k in range(len(distrs)):
        distr = distrs[k]
        name = names[k]
        param = params[k]
        population = np.zeros(length//step + 1) # all people that are in i generation
        active = np.zeros(length//step + 1) # people in i generation,that are finding parents (roots of processes)
        len_arr = length//step + 1
        for proc in processes:
            for i in range(len(proc.generations)) :
                gen = proc.generations[i]
                for p in gen:
                    if p.gender != 'male':
                        continue
                    bday = get_year(p.birthday)
                    if bday > -5000 :
                        population[(bday- start_time)//step ] += 1
                        if i == 0:
                            active[(bday- start_time)//step] += 1

        alone_years = []
        for i in range(len_arr-1,0,-1):
            active_cur = active[i]
            while(active_cur > 0) :
                offspring = distr(param,sz=1)
                active_cur -= offspring
                population[i-1] += 1
                active[i-1] += 1
            if population[i] == 1:
                alone_years.append(i)
        OY = [life_coef for x in alone_years]
        plt.plot(start_time + step*np.arange(len_arr),life_coef*population,label=name)
        plt.plot(start_time + step*np.array(alone_years),
                 OY,'*',label='only one man')
        if len(distrs) == 1 :
            plt.plot(start_time + step*np.arange(len_arr),
                     life_coef*active,'--',label='active roots')
    plt.legend(loc='best')
    plt.show()
```

Разделим всех мужчин на активных и не активных, активные - находящиеся в последнем поколении процесса, а также, удовлетворяющие (*) - критерию(см. пункт 3). Тогда каждый активный мужчина с датой рождения $b\_year_i$ будет порождать $\xi_i \sim distr$ активных мужчин c датами рождения $b\_year_j = b\_year_i + avg\_dist$ , где $avg\_dist$ - среднее расстояние между двумя поколения. Будем считать за одно $n$-ое поколение  мужчин с датой рождения $b\_year \in [avg\_dist \cdot n,avg\_dist \cdot (n+1) - 1]$

Для подсчета количества людей живущих в данном поколении (а не только активных) умножим количество активных на среднее число поколений живущих в данный отрезок времени.

```python
def calc_forward_all(distrs,names, params):
    first = 0
    last = 2400
    colors = ['r','g','b']
    plt.figure(figsize=(12,6))
    plt.grid(True)
    plt.xlabel('years')
    plt.ylabel('population')

    plt.xlim(1700,2400)
    for j in range(len(distrs)):
        distr = distrs[j]
        name = names[j]
        param = params[j]
        active = np.zeros((last+1)//step + 1)
        population = np.zeros((last+1)//step + 1)

        for proc in processes:
            for i in range(len(proc.generations)) :
                gen = proc.generations[i]
                for p in gen:
                    if p.gender != 'male':
                        continue
                    bday = get_year(p.birthday)
                    if bday > -5000 :
                        population[bday//step] += 1
                        if i == len(proc.generations) - 1 :
    #                     or (len(p.children) == 0 and death_crit(p)):
                            active[bday//step] += 1
        base_population = copy.copy(population)
        for i in range(len(active) - 1):
            offspring = np.sum(distr(param,active[i]))
            active[i+1] += offspring
            population[i+1] += offspring
        if j == 0 :
            plt.plot(step*np.arange((last+1)//step + 1), 
                     base_population,'--',color='black',label='base')
        plt.plot(step*np.arange((last+1)//step + 1),life_coef * population,
                 color=colors[j],label=name+'('+str(param)+')'+'reconstructed')
        
    plt.legend(loc = 'best',fontsize=12)
    plt.show()
```

Для каждого процесса в отдельности:
 - будем следить только за большими родами, в которых $\geq$ 6 поколений
 - не будем рассматривать поколения у которых $\lambda$ > 1.6 , т.к. они будут размножаться слишком быстро, и наверное об этом процессе просто очень мало данных.


```python
def calc_forward_processes(distr,params):
    first = 0
    last = 2600
    step = avg_gen_distance
    colors = ['r','g','b']
    population = np.zeros((last+1)//step + 1)
    num_procceses = np.zeros((last+1)//step + 1)
    
    for i in range(len(processes)):
        distribution = distr
        proc = processes[i]
        proc_existence = np.zeros((last+1)//step + 1)
        param = params[i]
        if param < 0 or len(proc.generations) < 6 or param > 1.6:
            continue
        active = np.zeros((last+1)//step + 1)
        first_active = -1
        for j in range(len(proc.generations)) :
            gen = proc.generations[j]
            for p in gen:
                if p.gender != 'male':
                    continue
                bday = get_year(p.birthday)
                if bday > -5000 :
                    population[bday//step] += 1
                    if j == len(proc.generations) - 1 :
                        active[bday//step] += 1
                        proc_existence[bday//step] = 1
                        if first_active == -1:
                            first_active = bday//step 
        if first_active == -1 :
            continue
        for y in range(first_active, len(active) - 1):
            if active[y] == 0 :
                break
            offspring = np.sum(distribution(param,active[y]))
            active[y+1] += offspring
            population[y+1] += offspring
            proc_existence[y+1] = 1
       
        num_procceses += proc_existence
            
    plt.figure(figsize=(12,6))
    plt.grid(True)
    plt.xlim(1500,last)
    plt.title('population(years)')
    plt.xlabel('years')
    plt.ylabel('population')
    plt.plot(step*np.arange((last+1)//step + 1), life_coef*population, color='g')
    plt.show()
    
    plt.figure(figsize=(12,6))
    plt.grid(True)
    plt.xlim(1600,last)
    plt.title('num of families(years)')
    plt.xlabel('years')
    plt.ylabel('families')
    plt.plot(step*np.arange((last+1)//step + 1), num_procceses, color='r')
```

Вероятность вырождения процессов(или конкретного процесса) на основе уже известных данных $q = q^{n}$ , где n - количество листьев с неизвестными потомками (т.к. все процессы независимы в совокупности).

Для нахождения всех значений, воспользуемся доказанной сходимостью производящей функции.

```python
def f(z,lmb):
    return np.exp(lmb*(z-1))

def get_q(lmb) :
    x = 1e-14
    for i in range(int(1e2)):
        x = f(x,lmb)
    return x

q_all = []
for j in range(len(processes)):
    proc = processes[j]
    lv = 0
    if pois_lmb[j] < 0 : 
        continue

    q = get_q(pois_lmb[j])
    for j in range(len(proc.generations)):
        gen = proc.generations[j]
        for p in gen:
            if p.gender != 'male':
                continue;
            if j == len(proc.generations) - 1 or get_year(p.birthday) > 1960:
                lv += 1
    q_all.append(q**lv)
print(np.mean(q_all))
```
